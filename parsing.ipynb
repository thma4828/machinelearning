```python
import base64
```

```python
with open("opedassign.pdf", "rb") as pdf_file:
    encoded_string = base64.b64encode(pdf_file.read())
```

```python
bytes_ = base64.decodebytes(encoded_string)
encoded_string.decode('cp1252')
```

```python
z = base64.b64encode(bytes_).decode('ascii')
```

```python
char_table = []
for ch in list(encoded_string):
    char_table.append(ord(chr(ch)))
```

```python
char_table
```

```python
ord('J')
```

```python
bin(5)
```

```python
bstrings = []
for c in char_table:
    binaries = bin(c)
    bstrings.append(binaries[2:])
bstrings
```

```python
concat_string = ''
for b in bstrings:
    concat_string += b
#concat_string
```

```python
new_bytes = []
counter = 0
prev_ind = 0
while counter < 100:
    if counter != 0 and (counter+1) % 8 == 0:
        new_bytes.append(concat_string[prev_ind:counter+1])
        prev_ind = counter + 1
    counter = counter + 1        
```

```python
new_bytes
```

```python
new_ints = []
for b in new_bytes:
    new_ints.append(chr(int(b, 2)))
```

```python
new_ints
ns = ''
for n in new_ints:
    ns +=  n  
ns
base64.standard_b64decode(z)
```
